```python
# 从继承到多继承
class people:
    # 定义基本属性
    name = ''
    age = 0
    # 定义私有属性
    __weight = 0
    # 定义构造方法
    def __init__(self,n,a,w):
        self.name = n
        self.age = a
        self.__weight = w
    def speak(self):
        print("%s 说: 我 %d 岁。" %(self.name,self.age))

# 单继承
class student(people):
    grade = ''
    def __init__(self,n,a,w,g):
        # 调用父类的构函
        people.__init__(self,n,a,w)
        self.grade = g
    # 重写父类的方法
    def speak(self):
        print("%s 说: 我 %d 岁了，我在读 %d 年级" %(self.name,self.age,self.grade))

# s = student('ken',10,60,6)
# s.speak()


# 再来一个类
class sperker():
    topic = ''
    name = ''
    def __init__(self,n,t):
        self.name = n
        self.topic = t
    def speak(self):
        print("我叫 %s，我是一个演说家，我演讲的主题是 %s"%(self.name,self.topic))

# 多继承

class sample(sperker,student):
    a = ''
    def __init__(self,n,a,w,g,t):
        student.__init__(self,n,a,w,g)
        sperker.__init__(self,n,t)
s = sample("ken",10,60,6,"python")
s.speak()
```

```python
# 方法重写
class Parent:
    def my_method(self):
        print('调用父类方法')

class Child(Parent):
    def my_method(self):
        print('调用子类方法')

c = Child()#子类实例
c.my_method()#调用子类已经重写的方法
super(Child,c).my_method()#调用父类已被覆盖的方法
```

```python
# 私有属性
class JustCounter:
    __secretCount = 0  # 私有变量
    publicCount = 0    # 公开变量

    def count(self):
        self.__secretCount += 1
        self.publicCount += 1
        
c = JustCounter()
c.count()
print(c.publicCount)
c.count()
print(c.publicCount)
print(c.publicCount)
print(c.__secretCount)#执行就会报错，实例不能访问私有变量

```

```python
# 私有方法
class Site:
    def __init__(self,name,url):
        self.name = name
        self.__url = url

    def who(self):
        print('name  :', self.name)
        print('url   :', self.__url)
    def __foo(self):
        print('这是一个私有方法')
    def foo(self):
        print('这是一个公共方法')
        self.__foo()

s = Site('菜鸟教程','www.runoob.com')
s.who()
s.foo()
s.__foo()#执行就会报错，实例不能访问私有方法
```

```python
# 运算符重载（Operator Overloading）
# 运算符重载就是给自定义类 “赋予运算符的意义”，让代码更直观、更像自然语言
class Vector:
    def __init__(self, a, b):
        self.a = a  # 向量的x分量
        self.b = b  # 向量的y分量

    def __str__(self):
        # 重载了字符串打印行为：print(v1) 时显示为 Vector (x, y)
        return 'Vector (%d, %d)' % (self.a, self.b)

    def __add__(self, other):
        # 重载了 + 运算符：两个向量相加时，分量分别相加
        return Vector(self.a + other.a, self.b + other.b)

v1 = Vector(2, 10)
v2 = Vector(5, -2)
print(v1 + v2)  # 输出 Vector (7, 8)
```
