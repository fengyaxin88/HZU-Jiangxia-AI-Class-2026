```python
X = torch.arange(12,dtype=torch.float32).reshape((3,4))
Y = torch.tensor([[2.0,1,4,3],[3,4,5,6],[1,2,3,4]])
torch.cat((X,Y),dim=0)  #在行上将两个张量连接起来
```

```python
X == Y , X > Y , X < Y
```

```python

X.sum() 
```

```python
#广播机制
a = torch.arange(3).reshape((3, 1))
b = torch.arange(2).reshape((1, 2))
a, b
```

```python
a + b
```

```python
#节省内存
Z = torch.zeros_like(Y)  
print('id(Z):', id(Z))  
Z[:] = X + Y          
print('id(Z):', id(Z))
```

```python

before = id(X)
X += Y          #使用原地操作符（如+=）将结果分配给先前分配的数组
id(X) == before
```

```python
#将张量转换为numpy中的张量
A = X.numpy()
B = torch.tensor(A)
C = B.numpy()
type(A), type(B),type(C)
```

```python
import os

# 1. 定义数据存储目录和文件名（
data_dir = 'D:'  
os.makedirs(data_dir, exist_ok=True)  
data_file = os.path.join(data_dir, 'student_scores.csv')  

# 2. 写入CSV数据
with open(data_file, 'w', encoding='utf-8') as f:   
    f.write('StudentID,Name,Math,English,Class\n')
    f.write('001,张三,95,88,高一(1)班\n')
    f.write('002,李四,82,NA,高一(2)班\n')
    f.write('003,王五,NA,90,高一(1)班\n')
    f.write('004,赵六,78,85,NA\n')
    f.write('005,钱七,92,89,高一(3)班\n')
#数据加载
import pandas as pd
data = pd.read_csv(data_file)
```

```python
#处理数据
#   插值法（ .fillna( . ) ）
inputs, outputs = data.iloc[:, 0:2], data.iloc[:,2]
inputs = inputs.fillna( inputs.mean(numeric_only=True) )
print(inputs)

inputs = pd.get_dummies(inputs, dummy_na=True)
inputs = inputs.astype(int)
print(inputs)
```

```python
#   数值型dataframe转化为张量
import torch
x,y = torch.tensor(inputs.values), torch.tensor(outputs.values)
print(x,y,sep="\n")

# 将大小为1的张量转换为Python标量
a = torch.tensor([3.5]) # 创建一个大小为1的张量
result = (a, a.item(), float(a), int(a)) # 多种方式将其转为Python标量
```
