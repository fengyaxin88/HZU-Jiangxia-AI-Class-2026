张量Tensor

```python
import torch 
```

```python
x=torch.arange(12).reshape(3,4)
y=torch.tensor([[3,4,5,6],[6,7,8,9],[9,10,11,12]],dtype=torch.float32)
x,y
```

```python
c1=torch.cat((x,y),dim=0)
c2=torch.cat((x,y),dim=1)
c1,c2
```

```python
sum1=x.sum()
sum2=x.sum(dim=0)
sum3=x.sum(dim=1)
sum1,sum2,sum3
```
