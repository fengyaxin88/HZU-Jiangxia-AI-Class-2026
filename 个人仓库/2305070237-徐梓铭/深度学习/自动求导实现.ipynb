```python
import torch
x=torch.arange(4.0)
x
```

```python
x.requires_grad_(True)#存储梯度
x.grad
```

```python
y=2*torch.dot(x,x)
y
```

```python
#调用反向传播函数
y.backward()
x.grad
```

```python
#验证
x.grad==4*x 
```

```python
#默认情况下会累计梯度，需要清除之前的值
x.grad.zero_()
y=x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y=x*x
y.sum().backward()
x.grad
```

```python
x.grad.zero_()
y=x*x
u=y.detach()
z=u*x
z.sum().backward()
x.grad==u
```
