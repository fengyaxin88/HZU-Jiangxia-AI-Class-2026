```python
# 给定具有相同形状的任意两个张量，任何按元素二元运算的结果都将是相同形状的张量
A = torch.arange(20,dtype=torch.float32).reshape(5,4)
B = A.clone()#创建A的一个副本
print(A,"\n",A+B)

# 矩阵的哈达玛积（Hadamard product）是按对应位置元素乘积得到的矩阵，与原先矩阵同形状
print(A*B)
```

张量求和

```python
t = torch.arange(4, dtype=torch.float32)
t, t.sum()#张量的求和
```

```python
print(A)
print(A.shape,A.sum())
```

 指定张量沿哪一个轴来通过求和降低维度
 可以按照降维的思想理解：
 axis=0降的就是行的维度，因此按照列来求和；
 axis=1降的是列的维度，按照行来求和

```python
# 1.按列求和
A_sum_axis0 = A.sum(axis=0)
A_sum_axis0, A_sum_axis0.shape
```

```python
# 2.按行求和
A_sum_axis1 = A.sum(axis=1) 
A_sum_axis1, A_sum_axis1.shape
```

```python
A.sum(axis=0).sum(axis=0)  # 先按行求和，再对结果按行求和
# 等价于
# A.sum()  # 直接对所有元素求和

# A是一个5X4的张量，只有0维和1维，按照所有维度求和就是全部求和
```

```python
# 一个与求和相关的量是平均值（mean或average）
print(A.mean())
print(A.sum()/A.numel())#numel返回张量中元素的数量,验证mean
```

```python
# 可以按照行或者列求和
A.mean(axis=0), A.sum(axis=0) / A.shape[0]
```

```python
# 参数keepdims=True用于计算总和或均值时保持轴数不变
sum_A = A.sum(axis=1, keepdims=True)
A,sum_A
```

```python
# 通过广播将A除以sum_A，这样就会得到每个元素占该列总和的比例
A/sum_A
```

```python
# 某个轴计算A元素的累积总和
# 按照列计算A的累积总和
print(A.cumsum(axis=0))
# 按照行计算A的累积总和
print(A.cumsum(axis=1))

#cumulative sum  |  cumsum
```

```python
# 点积是相同位置的按元素乘积的总和
y = torch.ones(4, dtype=torch.float32)
x, y, torch.dot(x, y)
```

```python
# 我们可以通过执行按元素乘法，然后进行求和来表示两个向量的点积
torch.sum(x * y)

# 注意区分两个矩阵的哈达玛积和点积，前者是按元素相乘得到一个矩阵，后者是按元素相乘后求和得到一个标量
```

```python
# torch.mv(A, x) 就是专门用来计算矩阵向量积的函数
# 矩阵向量积Ax是一个长度为m的列向量，其第i个元素是点积a⊤ix
print(A, x)
print(A.shape, x.shape, torch.mv(A, x))
```

```python
# 我们可以将矩阵-矩阵乘法AB看作是简单地执行m次矩阵-向量积，并将结果拼接在一起，形成一个n×m矩阵

B = torch.ones(4, 3)
B,torch.mm(A, B)

# 其实没必要挖牛角尖，就是线代里学到的矩阵乘法，这里从矩阵的哈达玛积，到矩阵-向量积，再推广到矩阵-矩阵积
# 就是为了引出矩阵乘法罢了
```

只有区分，torch.mv(A, x)是矩阵向量积，torch.mm(A, B)是矩阵矩阵乘法

三种范数

PyTorch 的 torch.norm() 函数默认计算的是 Frobenius 范数，且该范数只和矩阵元素的平方和有关，与矩阵的形状无关

```python
# L2范数是向量元素平方和的平方根：
u = torch.tensor([3.0, -4.0])
torch.norm(u)
```

```python
# L1范数，它表示为向量元素的绝对值之和：
torch.abs(u).sum()
```

```python
# 矩阵的弗罗贝尼乌斯范数（Frobenius norm）是矩阵元素平方和的平方根
torch.norm(torch.ones((4, 9)))
```

一点点奇怪的地方

```python
torch.ones((4, 9))
```

```python
torch.ones(4, 9)
```

torch.ones() 的官方定义是接收可变数量的整数参数，或者一个包含整数的序列（元组 / 列表） 来指定形状：
直接传多个整数：torch.ones(4, 9) → 把 4、9 分别识别为第一维、第二维的大小，对应形状 (4, 9)。
传一个元组：torch.ones((4, 9)) → 把元组 (4, 9) 解析成维度大小，结果也是形状 (4, 9)。
