```python
X=torch.arange(12,dtype=torch.float32).reshape(3,4)#多个张量结合
Y=torch.tensor([[2.0,1,4,3],[1,2,3,4],[4,3,2,1]])
torch.cat((X,Y),dim=0),torch.cat((X,Y),dim=1)#行和列合并
```

```python
#通过逻辑运算符构建二维张量
X==Y
```

```python
X.sum()
```
