存储梯度

```python
import torch
x=torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad
```

```python
y=2*torch.dot(x,x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad==4*x
```

```python
x.grad.zero_()#梯度清零
y=x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()#梯度清零
y=x*x
y.sum().backward()
x.grad
```

移动到计算图之外

```python
x.grad.zero_()#梯度清零
y=x*x
u=y.detach()
z=u*x
z.sum().backward()
x.grad==u
```

```python
x.grad.zero_()#梯度清零
y.sum().backward()
x.grad==2*x
```
