```python
import torch
```

```python
w = torch.tensor([3.0], requires_grad=True)
y = w**4 + 2*w**3
y.backward()
print(w.grad) 
```

```python
x = torch.linspace(-2, 2, 5, requires_grad=True)
f = torch.sin(x**2)  
f.sum().backward()
print(x.grad)
x.grad.zero_() 
print(x.grad)  
```

```python
x = torch.tensor([5.0], requires_grad=True)
f = torch.relu(x-3)
f.backward(retain_graph=True)  
(-f).backward()                
print(x.grad) 
```

```python
x = torch.tensor([4.0], requires_grad=True)
y = x**3
dy_dx = torch.autograd.grad(y, x)[0]
print(dy_dx)
```

```python
w = torch.ones(3, requires_grad=True)
w.grad = torch.tensor([10., 20., 30.])
optimizer = torch.optim.SGD([w], lr=0.1)
optimizer.step()
print(w)
```

```python
x = torch.tensor([2.0], requires_grad=True)
f = torch.log(1 + torch.exp(x))
grad = torch.autograd.grad(f, x, create_graph=True)[0]
grad2 = torch.autograd.grad(grad, x)[0]  # 二阶导
print(grad2)  # ≈0.10499 (σ(2)*(1-σ(2)))
```
